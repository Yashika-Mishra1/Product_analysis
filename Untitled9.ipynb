```python
#null value present in the data set so we have to we have to drop it 

nan_df = df[df.isna().any(axis=1)]
nan_df.head()
```

```python
dfa = df.drop_duplicates(subset= None,keep='first',inplace=False) #droping duplicate values
dfa
```

```python
#changing the name of head in column

dfa.columns = ['Order_ID','Product','Quantity_Ordered', 'Price_Each','Order_Date','Purchase_Address']
dfa.head()
```

```python
dfa['Date'] = dfa['Order_Date'].str[:8] #extracting date from order_date
dfa
```

```python
dfa['Month'] = dfa['Order_Date'].str[:2] #extracting month from order_date
dfa
```

```python
dfa[['House_No.','City','PIn_code']]= dfa.Purchase_Address.str.split(",",expand=True) #extracting House_No.,City,PIn_code from Purchase_Address.
```

```python
dfa.Month.value_counts() # to check the unique values # stucked here
```

```python
#here we can see that "or" value which is not a month so we are going to remove those rows which has or values in month
```

```python
dfa= dfa[dfa['Month'].str[:2] !='Or']
dfa.head()
```

```python
dfa.shape
```

```python
#data typecasting

dfa['Month'] = dfa['Month'].astype('int32')
dfa['Order_Date'] = pd.to_datetime(dfa['Order_Date'])
dfa['Date'] = pd.to_datetime(dfa['Date'])
dfa.info()
```

```python
dfa.Order_ID.value_counts()
```

```python
#orders are repeating so we cannot keep order id as index number
```

```python
dfa.Quantity_Ordered.value_counts()
```

```python
#changing the datatype

dfa['Quantity_Ordered']= dfa['Quantity_Ordered'].astype('int32')
dfa['Price_Each'] = dfa['Price_Each'].astype('float64')
dfa.info()
```

```python
#now lets make the sales column
```

# Ques - Find out the pair of products that are most sold together. For example generally, we would except a person who buys an iPhone 11 to also buy the Apple Airpods together with it. Identify such pairs of products and suggest what promotional offers should the retail company come up with based on your outcome?

```python
order = dfa[dfa['Order_ID'].duplicated(keep = False)] #ordering (filtering ) the data
order
```

```python
order['Ordered Together'] = order.groupby('Order_ID')['Product'].transform(lambda x : ','.join(x))
order = order[['Order_ID','Ordered Together']].drop_duplicates()
order.head(10)
```

```python
from itertools import combinations
from collections import Counter

count = Counter() #the counter object 'count' is used to count the occurrences of elements in the data_list

for row in order['Ordered Together']:
    items = row.split(',')
    count.update(Counter(combinations(items,2)))# most common pair of products bought together
    
for key,value in count.most_common(10): #top 10 products pair bought together
    print(key,value)                     
```

```python
# as we can see that iPhone and light charging cabel is the most sold product as a pair.
# This can be because company does not provide charger while selling the phone.
# second most sold product is usb cabel and google phone.
```
